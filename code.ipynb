# Summary of Count of Rows

```python
# Creating Categories
```

```python
neither = ((data['NNRTI'] == 0) & (data['PI'] ==0)).sum()
nnrti_only = ((data['NNRTI'] == 1) & (data['PI'] == 0) & (data['OTHER'] == 0)).sum()
pi_only = ((data['NNRTI'] == 0) & (data['PI'] == 1) & (data['OTHER'] == 0)).sum()
both = ((data['NNRTI'] == 1) & (data['PI'] == 1)).sum()
```

```python
# Showing counts of different drug combinations
```

```python
print('Given that NRTI is present,')

print('The count of Neither is: ', neither)
print('The count of NNRTI only is: ', nnrti_only)
print('The count of PI only is: ', pi_only)
print('The count of both NNRTI and PI is: ', both)
```

# Grouping rows by distinct CASEIDs to compute biomarker averages

```python
# specifying relevant biomarkers
```

```python
columns_to_average = ['CD4N', 'CD4P', 'CD8N', 'CD8P', 'CD3N', 'CD3P', 'HSRAT', 'CD4DT', 'HGB_LC', 'HCT_LC', 'MCV_LC', 'PLATLC', 'WBC_LC', 'RBC_LC', 'DIFFLC', 'LYMPLC', 'LPCTLC', 'LPMMLC', 'POLYLC', 'PPCTLC']
```

```python
averages_by_case_id = data.groupby('CASEID')[columns_to_average].mean().reset_index()
```
